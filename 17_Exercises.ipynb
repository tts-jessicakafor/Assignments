### Exercise 1. 

Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
````

````python
# Exercise 1

import pandas as pd

company_sales = pd.read_csv("https://raw.githubusercontent.com/tts-jessicakafor/tts-ds-fundamentals-course/main/datascience/python/11_mod_matplotlib/data/company_sales_data.csv")
# company_sales.describe()
company_sales
````

````python
plt.figure(figsize=(4,4), dpi=100)
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.plot(company_sales['month_number'],company_sales['total_profit'],color='blue')
ax = plt.gca()
ax.set_xlim([1, 12])
ax.set_ylim([150000, 450000])
plt.title("Total Profits of All months")
plt.show()
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# Exercise 3
````

````python
# facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer 	
plt.figure(figsize=(5,3), dpi=100)
plt.title("All Product Sales")
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
ax = plt.gca()
ax.set_xlim([1, 12])
ax.set_ylim([0, 14000])
plt.plot(company_sales['month_number'],company_sales['facecream'], label='Facecream', color='black', linestyle='dotted')
plt.plot(company_sales['month_number'],company_sales['facewash'], label='Facewash', color='blue')
plt.plot(company_sales['month_number'],company_sales['toothpaste'], label='Toothpaste', color='red')
plt.plot(company_sales['month_number'],company_sales['bathingsoap'], label='Bathing Soap', color='green',linestyle="dashdot")
plt.plot(company_sales['month_number'],company_sales['shampoo'], label='Shampoo', color='purple', marker = 'o')
plt.plot(company_sales['month_number'],company_sales['moisturizer'], label='Moisturizer', color='orange',linestyle="dashed",marker = 'v')
plt.legend()
plt.show()
````

As you can see there is not that much of a difference worth to use any visualizations. I decide to use the default selected style for the total units sold each month below. I used ylim method to set the max and minimum values and added labels. I have the rot=0 attribute to prevent the labels from turning.



````python
#https://www.tutorialspoint.com/automatically-setting-y-axis-limits-for-a-bar-graph-using-matplotlib
#https://stackoverflow.com/questions/64996074/how-to-use-column-values-for-x-axis-labels-in-matplotlib
````

````python
company_sales.plot.bar(x="month_number", y ="total_units", 
                    ylim=[company_sales['total_units'].min(), 
                    company_sales['total_units'].max()], rot=0)
plt.xlabel('Month')
plt.ylabel('Total Units')
plt.title('Total Units By Month')
plt.legend(('Units',),loc='upper center')
plt.show()
````
